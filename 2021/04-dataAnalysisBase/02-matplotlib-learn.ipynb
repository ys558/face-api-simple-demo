```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 60)
y = np.sin(x)
plt.figure(figsize=(8,4)) # 调出画布，8为长，4为高，单位是英寸（1英寸为25.4毫米）
plt.plot(x,y) # plot()的前两个参数分别代表X，Y轴数据的对象
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x**2)
plt.figure(figsize=(8,4))

plt.plot(x,y,label= "$sin(x)$", color="red",linewidth=2) # 对plot进行修饰：
plt.plot(x,z, "b--", label = "$cos(x^2)$")

plt.xlabel("Time(s)") # xlabel,ylabel：分别设置X，Y轴的标题文字:
plt.ylabel("V")

plt.title("pytho basis -- matplotlib") # itle：设置标题
plt.ylim(-1.2,1.2) # xlim,ylim：分别设置X,Y轴的显示范围
plt.legend(loc = 1) # legend:显示图例
plt.show()
```

```python
# 画直方图
a = np.random.randn(10000) # 产生符合标准正太分布的随机数
plt.hist(a, bins=60, edgecolor="black", alpha=0.7)
plt.show()
```
